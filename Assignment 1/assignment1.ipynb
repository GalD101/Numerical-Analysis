#### Part A:
$\text{In the lecture we saw that for tridiagonal matrices,}$  

$\text{the L part of the decomposition will only contain non-zero elements on the main diagonal and the sub-diagonal (one below the main diagonal).}$  

$\text{Similarly the U part will only contain non-zero elements on the main diagonal and the super-diagonal (one above the main diagonal)}$

```python
import numpy as np


# The regular decomposition from the exercise (O(n^3))
def lu_decomp(A: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Perform LU decomposition of a square matrix A using Doolittle's method.
    :param A (numpy.ndarray): an n X n (rows X columns) matrix.
    :return: L (numpy.darray): the lower triangular matrix of the LU decomposition with unit diagonal.
             U (numpy.darray): the upper triangular matrix of the LU decomposition.
    """

    n = len(A)
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    # Performing the decomposition
    for i in range(n):

        # Upper triangular matrix U
        for k in range(i, n):
            U[i][k] = A[i][k] - sum(L[i][j] * U[j][k] for j in range(i))

        # Lower triangular matrix L
        for k in range(i, n):
            if i == k:
                L[i][i] = 1  # Diagonal elements of L are set to 1
            else:
                L[k][i] = (A[k][i] - sum(L[k][j] * U[j][i] for j in range(i))) / U[i][i]
    return L, U


def lu_decomp_tridiagonal(A: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Performs an efficient O(n) LU decomposition of a tridiagonal matrix A.
    :param A (numpy.darray): an n X n (rows X columns) tridiagonal matrix.
    :return: L (numpy.darray): the lower triangular matrix of the LU decomposition with unit diagonal.
             U (numpy.darray): the upper triangular matrix of the LU decomposition.
    """

    n = len(A)
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    # this loop is running in O(n)
    for i in range(n):
        U[0, i] = A[0, i]  # first row of U is the same as the first row of A
        L[i, i] = 1  # Diagonal elements of L are set to 1

    # The rules for setting L and U came from solving it by hand and looking for ways to fix the issues I had on the internet
    # this loop is running in O(n)
    for j in range(1, n):
        L[j, j - 1] = A[j, j - 1] / U[j - 1, j - 1]

        if j < n:
            U[j - 1, j] = A[j - 1, j]

        U[j, j] = A[j, j] - L[j, j - 1] * U[j - 1, j]
    return L, U
```

```python
def forward_substitution_solve(L: np.ndarray, b: np.ndarray, N: int) -> np.ndarray:
    y = np.zeros(N)
    for i in range(N):
        sum_term = sum(L[i, j] * y[j] for j in range(i))
        y[i] = (b[i] - sum_term) / L[i, i]
    return y


def backward_substitution_solve(U: np.ndarray, b: np.ndarray, N: int) -> np.ndarray:
    x = np.zeros(N)
    for i in range(N - 1, -1, -1):
        sum_term = sum(U[i, j] * x[j] for j in range(i + 1, N))
        x[i] = (b[i] - sum_term) / U[i, i]
    return x
```

```python
def random_walker_time(N, p, n):
    t = 0;
    while n > 0:
        t = t + 1
        x = np.random.rand()
        if x < p:
            n = n + 1
        else:
            n = n - 1
        if n == N:
            n = N - 1
    return t
```

```python
N = 10
p = 0.55
q = 1 - p
possible_locations = [i for i in range(N)]
```

```python
# sparse does not work with python 3.13 :( (this is stupid) - solution: use a virtualenv with older version of python.
import sparse

# The matrix we get from the set of equations given in the assignment is tridiagonal, making it very sparse
identity = np.eye(N)
result = np.ones(N)

# Create the initial sparse matrix (I constructed it on paper then played around to get the same representation here
A = sparse.COO(identity + np.diag(-q * np.ones(N - 1), k=-1) + np.diag(-p * np.ones(N - 1), k=1) + [
    [p if j == 0 and i == 1 else -p if j == N - 1 and i == N - 2 else 0 for i in range(N)] for j in range(N)])

# for N = 4 for example, the solution vector is b = (0, 1, 1, 0)^T
b = np.asarray([0] + [1] * (N - 2) + [0])
A_dense = A.todense()

# Solve A*t = b using the methods I implemented for LU decomposition
L, U = lu_decomp_tridiagonal(A_dense)
y = forward_substitution_solve(L, b, N)
t = backward_substitution_solve(U, y, N)

# Monte Carlo simulation:
num_of_simulations = 10_0000

total_num_of_steps = [0 for _ in range(N)]
avg_num_of_steps = [0 for _ in range(N)]

for n in range(N):
    for _ in range(num_of_simulations):
        total_num_of_steps[n] += random_walker_time(N, p, n)
    avg_num_of_steps[n] = total_num_of_steps[n] / num_of_simulations

# plot the 2 results in the same graph, with a legend
plt.plot(possible_locations, avg_num_of_steps, label='Monte Carlo simulation')
plt.plot(possible_locations, t, label='LU decomposition')
plt.xlabel('n')
plt.ylabel(r'$T_{n}$')
plt.title('Average number of steps to get to point 0 as a function of initial condition n (location)')
plt.legend()
plt.show()
```
