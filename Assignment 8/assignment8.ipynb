$ A_k(t_{n+1})e^{ikx_j} = A_k(t_n)e^{ikx_j} + 1/2 r(A_k(t_n)e^{ikx_{j+1}} - A_k(t_n)e^{ikx_{j-1}}) $

$ \text{Dividing by } e^{ikx_j}: $

$ \text{Since } e^{ik\Delta x} - e^{-ik\Delta x} = 2 * i * \sin(k\Delta x): $

$ A_k(t_{n+1}) = A_k(t_n)[1 + ir sin(k\Delta x)] $

$ \text{And thus: } G = 1 + ir sin(k\Delta x) $

$ |G| = |1 + ir sin(k\Delta x)| = 1 + r^2 sin^2(k\Delta x) \leq 1 $
$ \text{and this inequality is never satisfied since } r^2 sin^2(k\Delta x) > 0 $
$ \forall r \neq 0 \in \mathbb{R}$

### Question 3

$ -5 \le x \le 5 \text{ and }  t = 0 ... 8$

$ \Delta x = 0.01, c = 1, $
$ y_0(x) =
  \begin{cases}
    1 & x > 4 \\
    0 & else
  \end{cases} $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.01
c = 1
T = 8
x_lim = [-5, 5]
x = np.arange(x_lim[0], x_lim[1] + 0.5 * dx, dx)

y0 = np.zeros(x.shape[0])
y0[x > 4] = 1

rs = [0.5, 1, 1.01]
```

#### Part A

Euler's method:

```python
def wave_euler(c, y0, T, dt, dx):
    r = c * dt / dx
    t = np.arange(0, T + 0.5 * dt, dt)
    y = np.zeros((t.shape[0], y0.shape[0]))
    y[0, :] = y0
    for n in range(len(t) - 1):
        y[n + 1, :-1] = r * y[n, 1:] + (1 - r) * y[n, :-1]
        y[n + 1, -1] = (1 - r) * y[n, -1]  # for the last x
    return y
```

```python
fig, axs = plt.subplots(1, len(rs), figsize=(15, 5))

for i, r in enumerate(rs):
    dt = r * dx / c
    y = wave_euler(c, y0, T, dt, dx)
    im = axs[i].imshow(y, extent=(x_lim[0], x_lim[1], 0, T), origin='lower')
    fig.colorbar(im, ax=axs[i])
    axs[i].set_title(f'Euler Method with r={r}')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('t')

plt.tight_layout()
plt.show()
```

#### Part B

Lax-Friedrichs method:

```python
def wave_lax_friedrichs(c, y0, T, dt, dx):
    r = c * dt / dx
    t = np.arange(0, T + 0.5 * dt, dt)
    y = np.zeros((t.shape[0], y0.shape[0]))
    y[0, :] = y0
    for n in range(len(t) - 1):
        y[n + 1, 1:-1] = 0.5 * ((1 + r) * y[n, 2:] + (1 - r) * y[n, :-2])
        y[n + 1, 0] = y[n + 1, 1]  # Neumann boundary (left boundary)
        y[n + 1, -1] = y[n + 1, -2]  # Neumann boundary (right boundary)
    return y
```

```python
fig, axs = plt.subplots(1, len(rs), figsize=(15, 5))

for i, r in enumerate(rs):
    dt = r * dx / c
    y = wave_lax_friedrichs(c, y0, T, dt, dx)
    im = axs[i].imshow(y, extent=(x_lim[0], x_lim[1], 0, T), origin='lower')
    fig.colorbar(im, ax=axs[i])
    axs[i].set_title(f'Lax-Friedrichs Method with r={r}')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('t')

plt.tight_layout()
plt.show()
```

#### Part C

Lax-Wendroff method:

```python
def wave_lax_wendroff(c, y0, T, dt, dx):
    r = c * dt / dx
    t = np.arange(0, T + 0.5 * dt, dt)
    y = np.zeros((t.shape[0], y0.shape[0]))
    y[0, :] = y0
    for n in range(len(t) - 1):
        y[n + 1, 1:-1] = (0.5 * r * (r + 1) * y[n, 2:] + (1 - r ** 2) * y[n, 1:-1] + 0.5 * r * (r - 1) * y[n, :-2])
        y[n + 1, 0] = y[n + 1, 1]  # Neumann boundary (left boundary)
        y[n + 1, -1] = y[n + 1, -2]  # Neumann boundary (right boundary)
    return y
```

```python
fig, axs = plt.subplots(1, len(rs), figsize=(15, 5))

for i, r in enumerate(rs):
    dt = r * dx / c
    y = wave_lax_wendroff(c, y0, T, dt, dx)
    im = axs[i].imshow(y, extent=(x_lim[0], x_lim[1], 0, T), origin='lower')
    fig.colorbar(im, ax=axs[i])
    axs[i].set_title(f'Lax-Friedrichs Method with r={r}')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('t')

plt.tight_layout()
plt.show()
```
